# Load data from files and create nicer plots


```python
# Imports
from PIL import Image
import matplotlib.pyplot as plt
import torch
import numpy as np
import os
from time import time
from scipy import stats
```

```python

# Load matrices
# https://vegibit.com/how-to-save-and-load-numpy-arrays/
samples_grid = np.loadtxt('./data/samples_grid.txt', delimiter=',')
samples_random = np.loadtxt('./data/samples_random.txt', delimiter=',')
samples_lhc = np.loadtxt('./data/samples_lhc.txt', delimiter=',')

iterations_grid = np.loadtxt('./data/iterations_grid.txt', delimiter=',')
iterations_random = np.loadtxt('./data/iterations_random.txt', delimiter=',')
iterations_lhc = np.loadtxt('./data/iterations_lhc.txt', delimiter=',')


# Put to dictionary 
iterations_convergence = {}
iterations_convergence['grid'] = iterations_grid
iterations_convergence['random'] = iterations_random
iterations_convergence['lhc'] = iterations_lhc


```

### Plot 1.

```python

iterations = np.arange(100, 20100, 500)

# PLT setting
plt.rcParams['figure.dpi'] = 400

plt.figure(figsize=(8, 4))
plt.plot(iterations, iterations_convergence['grid'], label='Grid sampling',linewidth=2,linestyle='--',color=(0.1, 0.1, 0.1, 0.3))
plt.plot(iterations, iterations_convergence['random'], label='Random sampling',linewidth=2,color=(0.125, 0.7, 0.03, 1)) #,color=(0.1, 1, 0.2, 0.8))
plt.plot(iterations, iterations_convergence['lhc'], label='Latin Hypercube',linewidth=2)

plt.xlim([iterations[0],iterations[len(iterations)-1]])
plt.ylim([1.5,1.55])
plt.xticks(iterations[::3])
#plt.xticks([100,1000,2500,5000,7500,10000,12500,15000,17500,19600])


plt.xlabel('Iterations')
plt.ylabel('Area')
#plt.title('Area convergence for different number of iterations and sampling techniques')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('plot1.png')

# plt.plot(iterations, computationtime_grid, label='grid')
# plt.plot(iterations, computationtime_random, label='random')
# plt.plot(iterations, computationtime_lhc, label='lhc')
""" plt.xlabel('Iterations')
plt.ylabel('Computationl time (s)')
plt.title('Computation time for calculating the area convergence for different number of iterations and sampling techniques')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show() """
```

### Plot 2

```python
areas_grid   = iterations_convergence['grid']
areas_random = iterations_convergence['random']
areas_lhc    = iterations_convergence['lhc']


plt.rcParams['figure.dpi'] = 400
plt.figure(figsize=(8, 4))

plt.plot(iterations, np.array(areas_grid)-areas_grid[-1], label='Grid sampling',linewidth=2,linestyle='--',color=(0.1, 0.1, 0.1, 0.3))
plt.plot(iterations, np.array(areas_random)-areas_random[-1], label='Random sampling',linewidth=2,color=(0.125, 0.7, 0.03, 1)) 
plt.plot(iterations, np.array(areas_lhc)-areas_lhc[-1], label='Latin Hypercube',linewidth=2)

plt.xlim([iterations[0],iterations[len(iterations)-1]])
plt.xticks(iterations[::3])

plt.xlabel('Iterations')
plt.ylabel('Area (j - i)')
#plt.title('Area convergence of Mandelbrot set with iterations ')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('plot2.png')
```
